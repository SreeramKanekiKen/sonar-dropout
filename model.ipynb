```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data=pd.read_csv('sonar_dataset.csv', header=None)
data.head()
```

```python
data.shape
```

```python
data.isna().sum()
```

```python
data.columns
```

```python
data[60].value_counts()
```

```python
X=data.drop(60, axis='columns')
y=data[60]
```

```python
X
```

```python
y=pd.get_dummies(y, drop_first=True)
```

```python
y
```

```python
y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2, random_state=10)
len(X_train)
```

```python
len(X_test)
```

```python
X_train.shape
```

```python
import tensorflow as tf
from tensorflow import keras

model=keras.Sequential([
    keras.layers.Dense(60, input_shape=(60,), activation='relu'),
    keras.layers.Dense(30, activation='relu'),
    keras.layers.Dense(15, activation='relu'),
    keras.layers.Dense(1, activation='sigmoid')
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

model.fit(X_train, y_train, epochs=100, batch_size=8)
```

```python
model.evaluate(X_test, y_test)
```

```python
model.predict(X_test[:5])
```

```python
y_pred=model.predict(X_test).reshape(-1)
print(y_pred[:10])

y_pred=np.round(y_pred)
print(y_pred[:10])
```

```python
y_test[:10]
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
dropout_model=keras.Sequential([
    keras.layers.Dense(60, input_shape=(60,), activation='relu'),
    keras.layers.Dropout(0.5),
    keras.layers.Dense(30, activation='relu'),
    keras.layers.Dropout(0.25),
    keras.layers.Dense(15, activation='relu'),
    keras.layers.Dropout(0.125),
    keras.layers.Dense(1, activation='sigmoid')
])

dropout_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

dropout_model.fit(X_train, y_train, epochs=100, batch_size=8)
```

```python
dropout_model.evaluate(X_test, y_test)
```

```python
dropout_y_pred=dropout_model.predict(X_test)
print(dropout_y_pred[:10])
```

```python
dropout_y_pred=dropout_y_pred.reshape(-1)
yp=np.round(dropout_y_pred)
print(yp[:10])
```

```python
y_test[:10]
```

```python
print(classification_report(y_test, yp))
```
